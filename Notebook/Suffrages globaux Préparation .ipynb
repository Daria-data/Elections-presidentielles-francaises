### 1965 Premier tour

```python
df=pd.read_csv('C:/Users/daria/Dropbox/Simplon/Programmation/Chef_d_oeuvre/Data/cdsp_presi1965t1_circ.csv',sep=',',encoding='utf-8')
df1965_1=df[['Code département','département','circonscription','Inscrits','Votants','Exprimés','Blancs et nuls']]
df1965_1=df1965_1.groupby(['Code département','département'], as_index=False).sum()
df1965_1.rename(columns={"circonscription":"Annee"},inplace=True)
df1965_1["Annee"]=1965
df1965_1=df1965_1.melt(id_vars=["Code département", "département","Annee"], 
        var_name="suffrage", 
        value_name="1_tour")
df1965_1 = df1965_1.sort_values(by=["Code département"])
df1965_1
```
